```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import numpy as np
from nuflux.detgeo import SimulateDetector, check_mem, plot_det
```

```python
sim = SimulateDetector(param = 'mutristan l', N_evals = 1e5)
```

```python
sim.run(Lss = 0, collision = 'mu+mu+', show_time = 0, show_components = 0)
```

```python
sim.event_timing(sec = 'nozzles')
sim.event_timing(sec = 'muon_detector_ec', fs = None)
sim.event_timing(sec = 'muon_detector_barrel', fs = None)
sim.event_timing(sec = 'hcal_ec', fs = None)
sim.event_timing(sec = 'hcal_barrel', fs = None)
sim.event_timing(sec = 'ecal_ec', fs = None)
sim.event_timing(sec = 'ecal_barrel', fs = None)
plt.legend(loc = 'upper left')
plt.yscale('linear')
```

```python
sim.plot(sec = 'all')
```

```python
fig, ax = plt.subplots()
_,y,z,_,times,_ = sim.get_data()
h = ax.scatter(z, y, c = times, zorder = 30,s = 3)
plot_det('det_v2', ax= ax)
fig.colorbar(h)
```
